# Konversi RGB ke HSV

```python
def rgb2hsv(gambar):
    out = np.empty_like(gambar)
    gambar = np.float32(gambar)  
    
    R = gambar[:,:,0]
    G = gambar[:,:,1]
    B = gambar[:,:,2]
    
    r = R/sum([R,G,B])
    g = G/sum([R,G,B])
    b = B/sum([R,G,B])

    V = np.maximum(r,g,b)
    
    S = 0 if V.all() == 0 else (1-(np.minimum(r,g,b)/V))

    if S.all() == 0:
        H = 0
    elif V == r:
        H = (60*(g-b)) / (S * V)
    elif V == g:
        H = 60 * (2 + ((b-r)/(S*V)))
    elif V == b:
        H = 60 * (4 + ((r-g)/(S*V)))

    H = H + 360 if H < 0 else H
    
    out[:,:,0] = H
    out[:,:,1] = S
    out[:,:,2] = V
    return out
```

```python
from skimage.color import rgb2hsv

hsv = rgb2hsv(gambar)
H = hsv[:, :, 0]
S = hsv[:,:,1]
V = hsv[:, :, 2]

fig, (ax0, ax1, ax2, ax3) = plt.subplots(ncols=4, figsize=(12, 8))

ax0.imshow(hsv)
ax0.set_title("HSV image")
ax0.axis('off')
ax1.imshow(H, cmap='hsv')
ax1.set_title("Hue")
ax1.axis('off')
ax2.imshow(S)
ax2.set_title("Saturation")
ax2.axis('off')
ax3.imshow(V)
ax3.set_title("Value")
ax3.axis('off')

fig.tight_layout()
```

# Konversi RGB ke HSI

```python
from math import sqrt,pow,pi,acos

def rgb2hsi(gambar):
    out = np.empty_like(gambar)
    gambar = np.float32(gambar)
    
    R = gambar[:,:,0]
    G = gambar[:,:,1]
    B = gambar[:,:,2]
    
    
    def theta(R, G, B):
        atas = ((R-G)+(R-B))/2    
        bawah = sqrt(pow((R-G),2) + ((R-B) * (G-B)))
        
        return acos(atas / bawah)
    
    def get_hue(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Hue '''
        R = R/360
        G = G/360
        B = B/360
        #mendapatkan contoh ukuran matrix
        H = R
        
        for b in range(len(R)):
            for k in range(len(R)):
                H[b][k] = theta(R[b][k], G[b][k], B[b][k])
                
                #jika citra B lebih besar dari citra G
                if B[b][k] > G[b][k]:
                    H[b][k] = ((360 * pi) / 180.0) - theta(R[b][k], G[b][k], B[b][k])
        
        return H
    
    def get_saturation(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Saturation '''
        
        #mendapatkan nilai terkecil dari citra RGB
        minimum = np.minimum(np.minimum(R, G), B)
        
        S = np.float32(1 - ((3 * minimum) / (R + G + B)))
        return S
    
    def get_intensity(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Intensity '''
        
        I = (R + G + B) / 3
        return I
    
    out[:,:,0] = get_hue(R,G,B)
    out[:,:,1] = get_saturation(R,G,B)
    out[:,:,2] = get_intensity(R,G,B)
    return out
```

```python
import os
import numpy
import cv2

def average(total, num):
    return total / num

imagename = "coklat.jpg"

# 
imagedata = cv2.imread(imagename)
imageheight, imagewidth = imagedata.shape[:2]
    
the_number_of_pixel = imageheight * imagewidth

# RGB 
r_value = imagedata[:,:,2]
g_value = imagedata[:,:,1]
b_value = imagedata[:,:,0]

# GRBHSV
hsvdata = cv2.cvtColor(imagedata, cv2.COLOR_BGR2HSV)

# HSV 
h_value = hsvdata[:,:,0]
s_value = hsvdata[:,:,1]
v_value = hsvdata[:,:,2]

# 
mean_b = average(b_value.sum(), the_number_of_pixel)
mean_g = average(g_value.sum(), the_number_of_pixel)
mean_r = average(r_value.sum(), the_number_of_pixel)
mean_h = average(h_value.sum(), the_number_of_pixel)
mean_s = average(s_value.sum(), the_number_of_pixel)
mean_v = average(v_value.sum(), the_number_of_pixel)

print("R = " + str(mean_r), "G = " + str(mean_g), "B = " + str(mean_b))
print("H = " + str(mean_h), "S = " + str(mean_s), "V = " + str(mean_v))
```
