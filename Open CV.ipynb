# Buka Gambar

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
gambar = cv2.imread('coklat.jpg')
plt.imshow(gambar)
plt.show()
```

# Feature RGB

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(12, 8))

ax0.imshow(gambar[:,:,2])
ax0.set_title("R")
ax0.axis('off')
ax1.imshow(gambar[:,:,1])
ax1.set_title("G")
ax1.axis('off')
ax2.imshow(gambar[:,:,0])
ax2.set_title("B")
ax2.axis('off')

fig.tight_layout()
```

```python
def get_rgb(gambar):
    gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)
    
    R = np.mean(gambar[:,:,0])
    G = np.mean(gambar[:,:,1])
    B = np.mean(gambar[:,:,2])
    
    return R,G,B
```

```python
r,g,b = get_rgb(gambar)
print('R\t: ',r)
print('G\t: ',g)
print('B\t: ',b)
```

# Feature HSV

```python
hsv = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(12, 8))

h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

ax0.imshow(h,cmap='hsv')
ax0.set_title("H")
ax0.axis('off')
ax1.imshow(s)
ax1.set_title("S")
ax1.axis('off')
ax2.imshow(v)
ax2.set_title("V")
ax2.axis('off')

fig.tight_layout()
```

```python
def get_hsv(gambar):
    hsv = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)
    H = np.mean(hsv[:,:,0])
    S = np.mean(hsv[:,:,1])
    V = np.mean(hsv[:,:,2])

    return H, S, V
```

```python
h,s,v = get_hsv(gambar)
print('H\t: ',h)
print('S\t: ',s)
print('V\t: ',v)
```
