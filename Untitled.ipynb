# Memisahkan Layer RGB

```python
import matplotlib.pyplot as plt

R = coklat.copy()
R[:,:,1] = 0
R[:,:,2] = 0

G = coklat.copy()
G[:,:,0] = 0
G[:,:,2] = 0

B = coklat.copy()
B[:,:,0] = 0
B[:,:,1] = 0


fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(12, 8))

ax0.imshow(R)
ax0.set_title("R")
ax0.axis('off')
ax1.imshow(G)
ax1.set_title("G")
ax1.axis('off')
ax2.imshow(B)
ax2.set_title("B")
ax2.axis('off')

fig.tight_layout()
```

# RGB

```python
mean_R = np.mean(R)
mean_G = np.mean(G)
mean_B = np.mean(B)

print("R : ",mean_R)
print("G : ",mean_G)
print("B : ",mean_B)
```

# RGB to HSV

```python
def rgb2hsv(R,G,B):
    r = R/sum([R,G,B])
    g = G/sum([R,G,B])
    b = B/sum([R,G,B])

    V = max(r,g,b)
    
    S = 0 if V == 0 else (1-((min([r,g,b]))/V))

    if S == 0:
        H = 0
    elif V == r:
        H = (60*(g-b)) / (S * V)
    elif V == g:
        H = 60 * (2 + ((b-r)/(S*V)))
    elif V == b:
        H = 60 * (4 + ((r-g)/(S*V)))

    H = H + 360 if H < 0 else H
    
    return H,S,V
```

```python
H,S,V = rgb2hsv(mean_R,mean_G,mean_B)
print("H : ",H)
print("S : ",S)
print("V : ",V)
```

# RGB to HSI

```python
from math import cos,acos,pow,sqrt,pi
def theta(R,G,B):
    '''Menghitung Sudut'''
    
    #rumus bagian atas
    atas = ((R-G)+(R-B))/2
    
    #rumus bagian bawah
    bawah = sqrt(pow((R-G),2) * (R-B) * (G-B))
    
    #rumus utama
    sudut = acos(atas/bawah) * 180/pi
    return sudut

def rgb2hsi(R,G,B):
    H = theta(R,G,B) if B <= G else (360-theta(R,G,B))
    S = 1 - ((3 * min(R,G,B))/sum([R,G,B]))
    I = sum([R,G,B]) / 3
    return H, S, I
```

```python
h,s,i = rgb2hsi(mean_R,mean_G,mean_B)
print("H : ",h)
print("S : ",s)
print("I : ",i)
```
