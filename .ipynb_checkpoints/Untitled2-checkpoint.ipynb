# Konversi RGB ke HSI

```python
from cv2 import merge
from math import sqrt,pow,pi,acos

def rgb2hsi(gambar):
    gambar = np.float32(gambar)
    
    R = gambar[:,:,2]
    G = gambar[:,:,1]
    B = gambar[:,:,0]
    
    
    def theta(R, G, B):
        atas = ((R-G)+(R-B))/2    
        bawah = sqrt(pow((R-G),2) + ((R-B) * (G-B)))
        
        return acos(atas / bawah)
    
    def get_hue(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Hue '''
        
        #mendapatkan contoh ukuran matrix
        H = R
        
        for b in range(len(R)):
            for k in range(len(R)):
                H[b][k] = theta(R[b][k], G[b][k], B[b][k])
                
                #jika citra B lebih besar dari citra G
                if B[b][k] > G[b][k]:
                    H[b][k] = ((360 * pi) / 180.0) - theta(R[b][k], G[b][k], B[b][k])
        
        return H
    
    def get_saturation(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Saturation '''
        
        #mendapatkan nilai terkecil dari citra RGB
        minimum = np.minimum(np.minimum(R, G), B)
        
        S = 1 - (3 / (R + B + B) * minimum)
        return S
    
    def get_intensity(R, G, B):
        ''' Fungsi Untuk Mendapatkan Nilai Intensity '''
        
        I = (R + G + B) / 3
        return I
    
    return merge((get_hue(R, G, B), get_saturation(R, G, B), get_intensity(R, G, B)))
```

# Konversi RGB ke HSV

```python
def rgb2hsv(gambar):
    
    gambar = np.float32(gambar)
    
    R = gambar[:,:,2]
    G = gambar[:,:,1]
    B = gambar[:,:,0]
    
    r = R/sum([R,G,B])
    g = G/sum([R,G,B])
    b = B/sum([R,G,B])

    V = np.maximum(r,g,b)
    
    S = 0 if V == 0 else (1-((np.minimum([r,g,b]))/V))

    if S == 0:
        H = 0
    elif V == r:
        H = (60*(g-b)) / (S * V)
    elif V == g:
        H = 60 * (2 + ((b-r)/(S*V)))
    elif V == b:
        H = 60 * (4 + ((r-g)/(S*V)))

    H = H + 360 if H < 0 else H
    
    return H, S, V
```
